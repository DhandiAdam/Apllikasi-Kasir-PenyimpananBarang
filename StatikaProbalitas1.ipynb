```python
import math

# Fungsi untuk menghitung rata-rata (mean)
def hitung_mean(data):
    total = sum(data)
    jumlah_data = len(data)
    rata_rata = total / jumlah_data
    return rata_rata

# Fungsi untuk menghitung simpangan rata-rata (standard deviation)
def hitung_simpangan_rata_rata(data, rata_rata):
    n = len(data)
    total_perbedaan_kuadrat = sum((x - rata_rata) ** 2 for x in data)
    simpangan_rata_rata = math.sqrt(total_perbedaan_kuadrat / (n - 1))
    return simpangan_rata_rata

# Data contoh
data = [98, 95, 92, 88, 87, 85, 81, 78, 73, 72, 65]

# Menghitung rata-rata
rata_rata = hitung_mean(data)

# Menghitung simpangan rata-rata
simpangan_rata_rata = hitung_simpangan_rata_rata(data, rata_rata)

# Menampilkan hasil
print("Data:", data)
print("Rata-rata:", rata_rata)
print("Simpangan Rata-rata:", simpangan_rata_rata)
```

```python
import numpy as np
from scipy.optimize import curve_fit

# Data penjualan
tahun = np.array([2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007])
penjualan = np.array([150, 140, 125, 110, 150, 156, 160, 168])

# Fungsi kuadrat untuk fitting
def fungsi_kuadrat(tahun, a, b, c):
    return a * tahun**2 + b * tahun + c

# Melakukan fitting menggunakan curve_fit
parameter, kovarian = curve_fit(fungsi_kuadrat, tahun, penjualan)

# Mendapatkan nilai parameter
a, b, c = parameter

# Hitung nilai tren
tren = a * tahun**2 + b * tahun + c

# Hitung nilai indeks musim
indeks_musim = penjualan / tren * 100

# Hitung perkiraan penjualan pada tahun 2008
tahun_2008 = 2008
perkiraan_2008 = a * tahun_2008**2 + b * tahun_2008 + c

# Tampilkan hasil
print("Tahun\tPenjualan (Ribu)\tTren")
for i in range(len(tahun)):
    print(f"{tahun[i]}\t{penjualan[i]}\t\t{tren[i]:.2f}")

print("\nIndeks Musim:", indeks_musim)
print(f"\nPerkiraan Penjualan pada Tahun 2008: {perkiraan_2008:.2f} Ribu")
```

```python
def prediksi_penjualan(tahun, penjualan):
    # Hitung jumlah Y dan XY
    sum_Y = sum(penjualan)
    sum_XY = sum(x * y for x, y in zip(tahun, penjualan))

    # Hitung a dan b
    a = sum_Y / len(penjualan)
    b = sum_XY / sum(x**2 for x in tahun)

    # Hitung tren
    tren = [a + b * x for x in tahun]

    # Hitung perkiraan penjualan pada tahun berikutnya
    tahun_terakhir = tahun[-1]
    perkiraan_tahun_terakhir = a + b * tahun_terakhir

    # Tampilkan hasil
    print("Tahun\tPenjualan (Ribu)\tTren")
    for i in range(len(tahun)):
        print(f"{tahun[i]}\t{penjualan[i]}\t\t{tren[i]:.2f}")

    print(f"\nPerkiraan Penjualan pada Tahun Terakhir: {perkiraan_tahun_terakhir:.2f} Ribu")

# Input data penjualan
tahun_input = list(map(int, input("Masukkan tahun (pisahkan dengan spasi): ").split()))
penjualan_input = list(map(int, input("Masukkan penjualan (pisahkan dengan spasi): ").split()))

# Panggil fungsi prediksi_penjualan dengan input yang diberikan
prediksi_penjualan(tahun_input, penjualan_input)

```
